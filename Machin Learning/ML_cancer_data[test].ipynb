### 1. 의사결정트리를 이용한 예측 정확도

```python
# cancer을 dataframe으로 변환
cancer_df = pd.DataFrame(data = cancer.data, columns = cancer.feature_names)
cancer_df
```

```python
# cancer_df에 target컬럼 추가
cancer_df['target'] = cancer.target
cancer_df.head()
```

```python
# 학습데이터와 테스트 데이터 세트로 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cancer_data,
                                                    cancer_label,
                                                    test_size = 0.2,
                                                    shuffle = False,
                                                    random_state = 100)


print('train data  :\n', X_train)
print('train label :\n', X_test)
print('test data   :\n', y_train)
print('test label  :\n', y_train)
```

```python
# 알고리즘을 활용한 학습과 분류
# 의사결정 트리 만들기
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(random_state = 100)
dt_clf
```

```python
# 학습시키기
dt_clf.fit(X_train, y_train)
```

```python
# 예측하기
pred = dt_clf.predict(X_test)

print('predict\n', pred)
print('y_test\n', y_test)
```

```python
# 예측 정확도 평가하기
from sklearn.metrics import accuracy_score

print('예측 정확도는 {}입니다.'.format(accuracy_score(y_test, pred)))
```

### 2. KFold를 이용한 교차검증

```python
from sklearn.model_selection import KFold
```

```python
# data load
fold_cancer = load_breast_cancer()
fold_cancer
```

```python
# 의사결정트리 만들기
features = fold_cancer.data
label = fold_cancer.target

fold_df_clf = DecisionTreeClassifier()
fold_df_clf
```

```python
# 10개의 폴더세트를 분리하여 각 폴드 세트별 정확도를 담을 리스트 생성
cv_accuracy = []
kfold = KFold(n_splits = 10)
```

```python
features.shape
```

```python
n_iter = 0

for train_idx, test_idx in kfold.split(features) :
    #print(len(train_idx), len(test_idx), train_idx, test_idx)
    X_train, X_val = features[train_idx], features[test_idx]
    y_train, y_val = label[train_idx], label[test_idx]
    
    # 학습 및 예측
    fold_df_clf.fit(X_train, y_train)
    pred = fold_df_clf.predict(X_val)
    n_iter += 1
    
    # 검증 정확도 측정
    acc = np.round(accuracy_score(y_val, pred), 4)
    print('{} 교차 검증 정확도 : {}, 학습 데이터 크기 : {}, 검증 데이터 크기 : {}'.format(n_iter, acc, X_train.shape[0], X_val.shape[0]))
    print('{} 검증 세트 인덱스 : {}'.format(n_iter, X_val.shape[0]))
    
    cv_accuracy.append(acc)
    
print(np.mean(cv_accuracy))
```

### 5. 의사결정나무 시각화

```python
import graphviz
from sklearn.tree            import DecisionTreeClassifier
from sklearn.datasets        import load_breast_cancer
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
from sklearn.tree import export_graphviz
```

```python
# max_depth를 제한둘 수 있음. 깊이가 너무 깊으면 과적합이 발생할 가능성이 높음
dt_clf = DecisionTreeClassifier(random_state=100, 
                                criterion='entropy', 
                                max_depth=4, 
                                min_samples_split=2)

X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, 
                                                    cancer_data.target, 
                                                    test_size=0.2, 
                                                    random_state=100)
```

### 6. GridSearchCV

```python
import graphviz
from sklearn.tree            import DecisionTreeClassifier
from sklearn.datasets        import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics         import accuracy_score

import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, 
                                                    cancer_data.target, 
                                                    test_size=0.2, 
                                                    random_state=100)
dt_clf = DecisionTreeClassifier(random_state=100)

grid_parameter = {'max_depth' : [1,2,3,4,5],
                  'min_samples_split' : [2,3]}
```

```python
# 튜닝
dt_searchCV = GridSearchCV(dt_clf, param_grid = grid_parameter, cv=10, refit=True)
```

```python
# 학습
dt_searchCV.fit(X_train, y_train)
```

```python
# 속성 확인
dt_searchCV.cv_results_
```

```python
# 속성들을 데이터프레임 형태로 변환
scoreDF= pd.DataFrame(dt_searchCV.cv_results_)
scoreDF
```

```python
# rank 순위 파악
scoreDF[['params', 'rank_test_score']]
```

```python
print('최적의 파라미터 : ',dt_searchCV.best_params_)
print('최적의 정확도   : ',dt_searchCV.best_score_)
print('최적의 Estimator: ',dt_searchCV.best_estimator_)
```

```python
# 최적의 Estimator를 이용한 예측
model = dt_searchCV.best_estimator_
pred = model.predict(X_test)

print('모델 정확도 : ',accuracy_score(y_test, pred))
```
