- filter() : 그룹객체에 filter()를 적용할 때 조건식을 가진 함수를 전달하면 참인 그룹만 남음

```python
titanic_datasets = sns.load_dataset('titanic')
```

```python
grp_class_titanic = titanic_datasets.groupby('class')
grp_class_titanic
```

```python
# data의 갯수가 200개 이상인 그룹만 필터링

def filter_func(x) :
    return len(x) >= 200
```

```python
grp_filter = grp_class_titanic.filter(filter_func)
grp_filter
```

```python
grp_filter = grp_class_titanic.filter(lambda x : len(x) >= 200)
grp_filter
```

```python
# age열의 평균이 > 30인 그룹만 필터링

age_filter = grp_class_titanic.filter(lambda x : x.age.mean() < 30)
age_filter
```

```python
# 나이에 대한 범주

titanic_datasets['age_qcut'] = pd.qcut(titanic_datasets['age'], 3, labels=['a','b','c'])
```

```python
# 성별, 선실, 나이그룹(age_qcut)에 의한 생존률을 데이터프레임으로 생성
# 행 : 성별, 나이그룹에 대한 다중인덱스
# 열 : 선실인덱스

resultDF = titanic_datasets.pivot_table(index = ['sex', 'age_qcut'],
                             columns = 'pclass',
                             values = 'survived',
                             margins = True,
                             margins_name = 'survived_ratio',
                             aggfunc = 'count')
```

```python
resultDF
```

```python
all_cnt = resultDF.loc['survived_ratio', 'survived_ratio'][0]
```

```python
survived_ration_df = resultDF / all_cnt
```
