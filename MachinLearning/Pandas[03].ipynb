### apply 함수
- lambda와 같이 사용
- 행이나 열 단위로 복잡한 데이터 가공이 필요한 경우
- apply 함수는 인자로 함수를 넘겨 받을 수 있다

```python
def get_square(a) :
    return a ** 2
```

```python
print('제곱근 : ', get_square(2))
```

```python
# 위 코드를 lambda로 변환
lambda_square = lambda a : a ** 2
```

```python
print('제곱근 : ', lambda_square(2))
```

```python
# embark_town의 문자열 갯수를 별도의 컬럼인 embark_len 컬럼으로 추가

datasets['embark_len'] = datasets['embark_town'].apply(lambda x : len(str(x)))
datasets[['embark_town', 'embark_len']]
```

```python
# age 피쳐를 이용하여 나이가 20세 미만이면 chid 그렇지 않으면 adult로 구분하여 child_adult 컬럼을 추가
# lambda if ~ else
# lambda 매개변수 : true if, false else

datasets['child_adult'] = datasets['age'].apply(lambda x : 'child' if x  < 20 else 'adult')
datasets[['age', 'child_adult']]
```

```python
datasets['child_adult_02'] = datasets.apply(lambda df : 'child' if df.age < 20 else 'adult', axis=1)
datasets
```

```python
# 15세 이하는 child ,  15~60 까지는 adult , 61 이상은 elderly 로 분류를해서 age_division 컬럼에 저장
# lambda 식으로 apply함수를 이용해서 구현한다면?

datasets['age_division'] = datasets.apply(lambda df : 'child' if df.age < 15 else ('adult' if df.age <= 60 else 'aelderly'), axis=1)
datasets
```

```python
def get_age_category(age) :
    category=""
    if age < 6 :
        category = 'baby'
    elif age < 12 :
        category = 'child'
    elif age < 19 :
        category = 'teenager'
    elif age < 24 :
        category = 'student'
    return category
```

```python
datasets['age_category'] = datasets['age'].apply(lambda age : get_age_category(age))
datasets['age_category'].value_counts()
```

```python
# 승객에 대한 나이(age)와 성별(sex)에 따른 카테고리를 'age/gender'로 추가
# 조건1. 20살이 넘으면 성별을 그대로 사용
# 조건2. 20살 미만이면 성별에 관계없이 'child'라고 정의하고 결측값이 있는 경우엔 'i don't know'라고 정의

condition01 = lambda df : df.sex if df.age >= 20 else 'child'
condition02 = lambda df : "i don't know" if pd.isnull(df.age) else condition01(df)
```

```python
datasets['age/gender'] = datasets.apply(condition02, axis=1)
datasets['age/gender'].value_counts()
```

```python
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
```

```python
df1
```

```python
df2
```

```python
result1 = pd.concat([df1,df2])
result1
```

```python
# ignore_index = True

result2 = pd.concat([df1,df2], ignore_index=True)
result2
```

```python
result3 = pd.concat([df1,df2], axis=1)
result3
```

```python
# join='inner' --> 교집합

result4 = pd.concat([df1, df2], axis=1, join='inner', ignore_index=True)
result4
```

```python
stock01 = pd.read_excel('../data/stock price.xlsx')
stock01
```

```python
stock02 = pd.read_excel('../data/stock valuation.xlsx')
stock02
```

```python
# 데이터 프레임 합치기 - 교집합
merge_inner = pd.merge(stock01, stock02, how='inner')
display(merge_inner)
# 데이터 프레임 합치기 - 합집합
merge_outer = pd.merge(stock01, stock02, on='id', how='outer')
display(merge_outer)
# 데이터 프레임 합치기 - 왼쪽 데이터프레임을 기준 , 키 값 분리
merge_left = pd.merge(stock01, stock02, left_on='stock_name', right_on='name', how='left')
display(merge_left)
# 데이터 프레임 합치기 - 오른쪽 데이터프레임을 기준 , 키 값 분리
merge_right = pd.merge(stock01, stock02, left_on='stock_name', right_on='name', how='right')
display(merge_right)
# 불린 인덱싱과 결합하여 원하는 데이터 찾아보자(price < 50000)
price = stock01[stock01['price']<50000]
display(price)
```
