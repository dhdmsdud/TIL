```python
cityDF = pd.read_excel('../data/시도별 전출입 인구수.xlsx')
```

```python
cityDF.head()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 서브셋 만들기

cityDF = cityDF.fillna(method='ffill')
mask = (cityDF['전출지별'] == '서울특별시') & (cityDF['전입지별'] != '서울특별시')
seoulDF = cityDF[mask]
```

```python
seoulDF
```

```python
# 전출지별 열 삭제

seoulDF = seoulDF.drop(['전출지별'], axis=1)
seoulDF
```

```python
# 원하는 열 인덱스를 다른 이름으로 변경

seoulDF = seoulDF.rename({'전입지별' : '전입지'}, axis=1)
seoulDF
```

```python
# 열을 인덱스로 바꾸기

seoulDF = seoulDF.set_index('전입지')
seoulDF
```

```python
# 경기도로 이동한 인구 데이터 값만 선택해서 시각화

datas = seoulDF.loc['경기도']
print(datas.index)
print(datas.values)
```

```python
plt.figure(figsize=(14, 5))

plt.plot(datas.index, datas.values, marker='o', markersize=5)
plt.title('서울 -> 경기 인구 이동', size=20) # 그래프 제목
plt.xlabel('년도', size=15) # x축 제목
plt.ylabel('이동 인구 수', size=15) # y축 제목
plt.xticks(rotation='vertical') # x축의 인덱스를 세로로 바꿈
plt.legend(labels=['서울 -> 경기'], loc='best') # 범례를 넣는 최적의 위치

plt.show()
plt.close()
```

```python
# subplot 만들기

fig = plt.figure(figsize=(10,10))
ax01 = fig.add_subplot(2,1,1)
ax02 = fig.add_subplot(2,1,2)

ax01.plot(datas, marker='*', markersize=10)
ax02.plot(datas, marker='o', markersize=10, color='red', linewidth=5, label='서울 -> 경기', markerfacecolor='green')
ax02.legend(loc='best')
ax01.set_ylim(50000, 700000) # y축 범위 지정(최댓값, 최솟값)
ax02.set_xticklabels(datas.index, rotation=70) # 기울기

plt.show()
plt.close()
```

```python
# 서울에서 충청남도, 경상북도, 강원도로 이동한 인구 데이터를 데이터 프레임 형태로 만들기

# loc사용
dataDF = seoulDF.loc[['충청남도', '경상북도', '강원도']]

# map사용
col_years = list(map(str, range(1970, 2018))) 
dataDF = seoulDF.loc[['충청남도', '경상북도', '강원도'], col_years]
dataDF
```

```python
fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1,1,1)

ax.plot(col_years, dataDF.loc['충청남도', :], marker='o', label='서울 -> 충청', markerfacecolor='olive', markersize=10)
ax.plot(col_years, dataDF.loc['경상북도', :], marker='o', label='서울 -> 경상', markerfacecolor='yellow', markersize=10)
ax.plot(col_years, dataDF.loc['강원도', :], marker='o', label='서울 -> 강원도', markerfacecolor='coral', markersize=10)
plt.title('서울 -> 충청, 경상, 강원도 인구 이동', size=20)
plt.xlabel('년도', size=15)
plt.ylabel('이동 인구 수', size=15)
ax.legend(loc='best')

plt.show()
plt.close()
```

- bar plot

```python
Y = [2,3,1]
X = np.arange(len(Y))
```

```python
plt.figure()
plt.bar(X,Y) # 수직
#plt.barh(X,Y) # 수평

plt.title('bar plot')
plt.xlabel('선실 등급')
plt.ylabel('생존자')

xlabel = ['1등급','2등급','3등급']
plt.xticks(X, xlabel)
plt.yticks(Y)

plt.show()
plt.close()
```

```python
# 서울에서 충청남도, 경상북도, 강원도, 전라남도로 이동한 인구 데이터의 프레임 만들기
# 2010 ~ 2017년 사이의 데이터

col_years = list(map(str, range(2010, 2018))) 
dataDF = seoulDF.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years].T
dataDF
```

```python
dataDF.index
```

```python
# 데이터프레임에서 바로 시각화

dataDF.index = dataDF.index.map(int)
dataDF.index
```

```python
plt.figure()

dataDF.plot(kind='bar', color=['cadetblue', 'lightsalmon', 'mediumturquoise', 'slategrey'])
plt.title('인구이동', size=20)
plt.xlabel('기간')
plt.ylabel('인구수')

plt.show()
plt.close()
```

```python
# 2010 ~ 2017년 이동인구 수 를 합계하여 새로운 열(합계)을 추가
# 가장 큰 값으로 정렬, 수평막대그래프 이용

dataDF = dataDF.T
dataDF
```

```python
dataDF['합계'] = dataDF.sum(axis=1)
```

```python
sumDF = dataDF[['합계']].sort_values(by='합계', ascending=True)
sumDF
```

```python
plt.figure()

sumDF.plot(kind='barh', color='peachpuff')
plt.title('인구이동', size=20)
plt.xlabel('인구이동 수')
plt.ylabel('전입지')

plt.show()
plt.close()
```

```python
import seaborn as sns
```

```python
iris_datasets = sns.load_dataset('iris')
iris_datasets.head()
```

```python
iris_datasets.rename(columns = {iris_datasets.columns[0] : 'SL',
                     iris_datasets.columns[1] : 'SW',
                     iris_datasets.columns[2] : 'PL',
                     iris_datasets.columns[3] : 'PW',
                     iris_datasets.columns[4] : 'Y',}, inplace=True)

iris_datasets.head()
```

```python
# 종을 기준으로 그룹화하여 각 그룹의 평균

iris_mean_df = iris_datasets.groupby('Y').mean()
iris_mean_df
```

```python
iris_mean_df.T
```

```python
plt.figure()

iris_mean_df.T.plot(kind='bar')

plt.title('각 변수별 평균')
plt.xlabel('변수')
plt.ylabel('평균')

plt.show()
plt.close()
```
