```python
from sklearn.metrics.pairwise import cosine_similarity

import pandas as pd
import numpy as np
```

### Item-based Collaborative Filtering

```python
# 선호 --> 2, 안좋아함 --> 1, 안해봄 --> 0
item_travel = pd.DataFrame({'A' : [1,2,2,1,1,2],
                       'B' : [2,2,2,2,1,2],
                       'C' : [1,1,2,1,2,0],
                       'D' : [0,1,2,2,1,0],
                       'E' : [2,0,0,1,1,2],
                       'F' : [2,0,1,2,1,1],
                       'G' : [1,0,2,0,1,0],
                       'H' : [0,1,0,2,0,1],
                       'I' : [0,2,0,1,0,2]}, 
                       index = ['전시회', '레포츠', '캠핑', '드라이브', '맛집탐방', '휴양'],
                       dtype = 'int')
item_travel
```

```python
similarity = pd.DataFrame(cosine_similarity(item_travel),
                          columns = ['전시회', '레포츠', '캠핑', '드라이브', '맛집탐방', '휴양'],
                          index = ['전시회', '레포츠', '캠핑', '드라이브', '맛집탐방', '휴양'])
similarity
```

### User-based Collaborative Filtering

```python
user_travel = item_travel.T
user_travel
```

```python
similarity = pd.DataFrame(cosine_similarity(user_travel),
                          columns = ['A','B','C','D','E','F','G','H','I'],
                          index = ['A','B','C','D','E','F','G','H','I'])
similarity
```

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.decomposition import TruncatedSVD


columns = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('c:/Users/ruby/data/ml-100k/u.data.csv', sep='\t', names=columns)


columns = ['item_id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
          'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror',
          'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

movies = pd.read_csv('../data/ml-100k/u.item', sep='|', names=columns, encoding='latin-1')
movie_names = movies[['item_id', 'movie title']]

combined_movies_data = pd.merge(df, movie_names, on='item_id')
combined_movies_data.head()
```

```python
combined_movies_data.shape
```

```python
rating_crosstab = combined_movies_data.pivot_table(values='rating', index='user_id', columns='movie title', fill_value=0)
rating_crosstab
```

```python
# item-based 이므로 transpose
X = rating_crosstab.T
```

```python
# 행렬을 축소하여 상위 몇개만 보여줌
SVD = TruncatedSVD(n_components=12, random_state=5)
resultant_matrix = SVD.fit_transform(X)
resultant_matrix.shape
```

```python
# correlation matrix
corr_mat = np.corrcoef(resultant_matrix)
corr_mat.shape
```

```python
# 토이스토리랑 비슷한 영화
col_idx = rating_crosstab.columns.get_loc("Toy Story (1995)")
corr_specific = corr_mat[col_idx]
pd.DataFrame({'corr_specific':corr_specific, 'Movies': rating_crosstab.columns}).sort_values('corr_specific', ascending=False)
```
