2. 데이터 확인 및 구조분석과 각 데이터 프레임에 헤더를 추가

- ratings.dat : userId, movieId, rating, timestamp
- movies.dat : movieId, title, genres
- users.dat(뒤에 두 열은 제거) : userId, gender, age

```python
ratingDF = pd.read_csv('../data/ratings.dat', sep='::', names=['userId','movieId','rating','timestamp'])
ratingDF
```

```python
movieDF = pd.read_csv('../data/movies.dat', sep='::', names=['movieId','title','genres'])
movieDF
```

```python
userDF = pd.read_csv('../data/users.dat', sep='::', names=['userId','gender','age','a','b']).drop(columns=['a','b'])
userDF
```

3. 공통 컬럼을 이용한 데이터 병합

```python
movie_dataset = pd.merge(ratingDF, movieDF, on='movieId')
movie_datasets = pd.merge(movie_dataset, userDF, on='userId')
movie_datasets
```

4. user 평점 갯수 순위 확인(sort)

```python
rating_cnt = movie_datasets['userId'].sort_values().value_counts()
rating_cnt
```

10. 평점이 220개 이상인 영화만 필터링

```python
avg_220 = movie_datasets.groupby('movieId')
```

```python
avg_220.size()
```

```python
avg_220_filter = avg_220.filter(lambda x : len(x) >= 220).sort_values(by='userId')
avg_220_filter
```
