- feature dataset 확인

```python
iris_data = iris.data
iris_data
```

- 레이블(결정값, 타겟, 클래스)확인

```python
iris_label = iris.target
iris_label
```

- dataframe 변환

```python
iris_df = pd.DataFrame(data = iris_data,
                       columns = iris.feature_names)

iris_df['target'] = iris_label

iris_df
```

- 학습데이터(training data : validation data)와 테스트 데이터(test data) 세트로 분리
- train_test_split(test_size, train_size, shuffle, random_size)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, shuffle=False, random_state=20)
print('train data  :\n', X_train)
print('train label :\n', X_test)
print('test data   :\n', y_train)
print('test label  :\n', y_train)
```

- 알고리즘을 활용한 학습과 분류

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_clf = DecisionTreeClassifier(random_state = 20)
dt_clf
```

```python
# 학습(fit)
dt_clf.fit(X_train, y_train)
```

- 예측은 학습데이터가 아닌 다른 데이터(test data)를 이용해야 한다.

```python
prediction = dt_clf.predict(X_test)
print('predict\n', prediction)
print('y_test\n', y_test)
```

- 예측 정확도 평가

```python
from sklearn.metrics import accuracy_score
print('예측 정확도 : {0:.2f}'.format(accuracy_score(y_test, prediction)))
```

- 학습 데이터와 테스트 테이터를 분리 시키지 않았을 때

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

bad_iris = load_iris()

train_data = bad_iris.data
train_label = bad_iris.target

bad_dt_clf = DecisionTreeClassifier(random_state=20)
bad_dt_clf.fit(train_data, train_label)

pred = bad_dt_clf.predict(train_data)
print('예측 정확도 : {0:.2f}'.format(accuracy_score(train_label, pred)))
```
